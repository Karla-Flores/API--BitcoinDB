```python
# Creating a list based on csv
cdata = [c2,c3,c4,c5,c6,c7,c8,c9,c10,c11,c12,c13,c14,c15,c16,c17,c18,c19,c20,c21,c22,c23,c24,c25,c26,c27,c28,c29,c30,c31,c32,c33,c34,c35]
```

```python
# Creating empty list
c_data = []

# Loop for appending csv into a data frame
for file in cdata:
    c_data.append(pd.read_csv(file))

# Concat all dataframes
cdata = pd.concat(c_data, sort=False)
cdata
```

```python
cdata.drop(['date_added', 'tags','last_updated_y', 'platform', 'slug','quote','last_updated_x'], axis = 1, inplace = True) 
```

```python
x = cdata['name']
x = x.drop_duplicates()
len(x)
```

```python
cdata['tradable']=0
```

```python
# creating a loop for filling the 'difference' column using loc in the symbol column then append into a dataframe
d=[]
for i in x:
    c = pd.DataFrame(cdata.loc[cdata.name == i, :])
    c['difference']=(c.price.diff(1)).round(decimals=5)
    c['difference'] = c['difference'].astype(float)


    c[['%_difference']] = c[['difference']].pct_change().round(decimals=5)
    c['%_difference'] = c['%_difference'].astype(float)

    d = pd.DataFrame(d.append(c))
```

```python
# filling n/a with 0
d.fillna(0, inplace=True)
d.drop(['name'], axis=1,inplace=True)
d.reset_index(drop=True, inplace=True)
```

```python
d = d.replace([np.inf, -np.inf], 0)
```

```python
d.to_csv('out.csv')
```
