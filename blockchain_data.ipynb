```python
# Creating a list based on csv
bdata = [d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12,d13,d14,d15,d16,d17,d18,d19,d20,d21,d22,d23,d24,d25,d26,d27,d28,d29,d30,d31,d32,d33,d34,d35,d36,d37,d38,d39,d40,d41,d42,d43,d44,d45,d46,d47,d48,d49,d50,d51,d52,d53]
```

```python
# Creating empty list
b_data = []

# Loop for appending csv into a data frame
for file in bdata:
    b_data.append(pd.read_csv(file))

# Concat all dataframes
bdata = pd.concat(b_data, sort=False)
bdata
```

```python
# filling n/a with 0
bdata.fillna(0, inplace=True)
```

```python
# getting list of symbols
df= bdata['symbol']
df = df.drop_duplicates()
len(df)
```

```python
# creating a loop for filling the 'difference' column using loc in the symbol column then append into a dataframe
d=[]
for i in df:
    c = pd.DataFrame(bdata.loc[bdata.symbol == i, :])
    c['difference']=(c.last_trade_price.diff(1)).round(decimals=5)
    c['difference'] = c['difference'].astype(float)


    c[['%_difference']] = c[['difference']].pct_change().round(decimals=5)
    c['%_difference'] = c['%_difference'].astype(float)
    d=pd.DataFrame(d.append(c))
```

```python
# filling n/a with 0
d.fillna(0, inplace=True)
d.reset_index(drop=True, inplace=True)
```

```python
d= d.reindex(columns=['symbol', 'price_24h', 'volume_24h', 'last_trade_price', 'difference', '%_difference', 'date' ])
d
```

```python
d = d.replace([np.inf, -np.inf], 0)
```
