```python
# Dependencies & set up
import pandas as pd
import numpy as np 
```

```python
# Getting fata from csv files
d1 = 'blockchain_11-05-2021.csv'
d2 = 'blockchain_11-06-2021.csv'
d3 = 'blockchain_11-07-2021.csv'
d4 = 'blockchain_11-08-2021.csv'
d5 = 'blockchain_11-09-2021.csv'
d6 = 'blockchain_11-10-2021.csv'
d7 = 'blockchain_11-11-2021.csv'
d8 = 'blockchain_11-12-2021.csv'
d9 = 'blockchain_11-13-2021.csv'
d10 = 'blockchain_11-15-2021.csv'
d11 = 'blockchain_11-16-2021.csv'
d12 = 'blockchain_11-17-2021.csv'
d13 = 'blockchain_11-18-2021.csv'
d14 = 'blockchain_11-19-2021.csv'
d15 = 'blockchain_11-20-2021.csv'
d16 = 'blockchain_11-21-2021.csv'
d17 = 'blockchain_11-22-2021.csv'
d18 = 'blockchain_11-23-2021.csv'
d19 = 'blockchain_11-25-2021.csv'
d20 = 'blockchain_11-26-2021.csv'
d21 = 'blockchain_11-29-2021.csv'
d22 = 'blockchain_11-30-2021.csv'
```

```python
# Creating a list based on csv
bdata = [d1, d2, d3, d4, d5, d6, d7, d8, d9, d10, d11, d12, d13, d14, d15, d16, d17, d18, d19 ,d20, d21,d22]
```

```python
# Creating empty list
b_data = []

# Loop for appending csv into a data frame
for file in bdata:
    b_data.append(pd.read_csv(file))

# Concat all dataframes
bdata = pd.concat(b_data, sort=False)
bdata
```

```python
# filling n/a with 0
bdata.fillna(0, inplace=True)
```

```python
# getting list of symbols
df= bdata['symbol']
df = df.drop_duplicates()
len(df)
```

```python
# creating a loop for filling the 'difference' column using loc in the symbol column then append into a dataframe
d=[]
for i in df:
    c = pd.DataFrame(bdata.loc[bdata.symbol == i, :])
    c['difference']=(c.last_trade_price.diff(1)).round(decimals=4)
    c['difference'] = c['difference'].astype(float)
    c['%_difference']=(c['difference']/100).round(decimals=4)
    c['%_difference'] = c['%_difference'].astype(float)
    d=pd.DataFrame(d.append(c))
```
